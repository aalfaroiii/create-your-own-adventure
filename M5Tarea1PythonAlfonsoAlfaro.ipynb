```python
## a. Con la técnica .loc[], seleccinamene todas las filas de las siguientes columnas:
 ## 'Player', 'Squad', 'Gls/90', 'Ast/90', 'xG/90', 'xA/90' (3 puntos)
df_laliga21.columns
  # ayuda total por parte de SantiNieto para remplazar las columnas de .1 a /90
df_laliga21.columns = [col.replace('.1', '/90') if col.__contains__('.1') else col for col in df_laliga21.columns]

cols = ['Player','Squad','Gls/90','Ast/90','xG/90','xA/90']

df_laliga21['Player'] = df_laliga21.Player.apply(lambda x:x.split ('\\') [0])
df_laliga21.loc[:, cols]
```

```python
import matplotlib.pyplot as plt

from soccerplots.radar_chart import Radar

# Configuración
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df_laliga21.columns
```

```python
def radar_chart(df_laliga21, player, col_name_player, col_name_team, cols):

  players = df_laliga21[df_laliga21[col_name_player] == player]
  values = players[cols].values[0]
  ranges = [(df_laliga21[param].min(), df_laliga21[param].max()) for param in cols]

  title = dict(
    title_name=player,
    title_color='black',
    title_fontsize=20,

    subtitle_name=players[col_name_team].values[0],
    subtitle_color='black',
    subtitle_fontsize=15
  )

  fig, ax = plt.subplots(figsize=(18, 15))
  radar = Radar(label_fontsize=20, range_fontsize=12)
  fig, ax = radar.plot_radar(ranges=ranges, params=cols, values=values, radar_color=['#ba4f45', '#FFFFFF'], figax=(fig, ax), title=title)
```

```python
cols = ['Gls/90', 'Ast/90', 'xG+xA', 'npxG', 'G-PK/90', 'Min']
```

```python
radar_chart(df_laliga21, 'Lionel Messi', 'Player', 'Squad', cols)
```

```python
def radar_chart_2(df_laliga21, player, player2, col_name_player, col_name_team, cols):

  pl1 = df_laliga21[df_laliga21[col_name_player] == player]
  val1 = pl1[cols].values[0]

  pl2= df_laliga21[df_laliga21[col_name_player] == player2]
  val2 = pl2[cols].values[0]

  values = [val1, val2]
  ranges = [(df_laliga21[col].min(), df_laliga21[col].max()) for col in cols]

  title = dict(
    title_name=player,
    title_color='#B6282F',
    subtitle_name=pl1[col_name_team].values[0],
    subtitle_color='#B6282F',

    title_name_2=player2,
    title_color_2='#344D94',
    subtitle_name_2=pl2[col_name_team].values[0],
    subtitle_color_2='#344D94',
    title_fontsize=18,
    subtitle_fontsize=15,
  )
  fig, ax = plt.subplots(figsize=(18, 15))
  radar = Radar(label_fontsize=20, range_fontsize=12)
  fig, ax = radar.plot_radar(ranges=ranges, params=cols, values=values, radar_color=['#B6282F', '#344D94'], figax=(fig, ax), title=title, compare=True)
```

```python
radar_chart_2(df_laliga21, 'Gerard Moreno', 'Iago Aspas', 'Player', 'Squad', cols)
```
